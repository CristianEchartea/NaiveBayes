```python
#LECTURA DE LOS ARCHIVOS
import nltk
#nltk.download('punkt')
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
from os import listdir



#subdirectorio 1
dirc='entornDataset';
sub_dirs=['enron4','enron2','enron6','enron1','enron3','enron5']
#sub_dirs=['enron4']

#CARGAMOS LOS CORREOS DE SPAM EN MEMORIA
def CargarCorreos(value,lista):
    for i in sub_dirs:
        for cosa in listdir(dirc+"/"+i+value):
            f= open(dirc+"/"+i+value+cosa,'r',encoding='utf-8', errors='ignore')
            #print(f.mode)
            texto= f.read()
            f.close();
            texto=limpiarTexto(texto)
            lista.append(texto);
            #print(cosa)


import re            
def limpiarTexto(texto):
    texto= texto.lower()
    stop_w= set(stopwords.words('english'))
    text_token= word_tokenize(texto)
            
    res=  [w for w in text_token if not w in stop_w]
    res=  [w for w in res if len(w)>3 and len(w)<15]
    res= ' '.join(res)
    vec=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','w','x','y','z']
    for i in vec:
        cadena=i+i+i+"*."
        eliminar_palabras= re.compile(cadena)
        res= eliminar_palabras.sub("",res);
    res= res.replace("[","");
    eliminar_palabra=re.compile("[0-9][0-9]*.")
    res= eliminar_palabras.sub("",res); 
    res= res.strip()
    return res
            
lista_spam=[];
CargarCorreos("/spam/",lista_spam);

                    

```
